```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('travel_insurance.csv')
```

```python
df
```

Here, we split the dataset to two parts: customers and non-customers.

```python
df_customers = df[df['TravelInsurance']==1]
df_non_customers = df[df['TravelInsurance']==0]
```

1. Age

```python
ax = sns.countplot(x='Age', data=df_customers)
ax.set_yticks([0,50,100,150,200,250,300,350,400])
plt.title('Distribution of age of customers')
```

```python
ax = sns.countplot(x='Age', data=df_non_customers)
ax.set_yticks([0,50,100,150,200,250,300,350,400])
plt.title('Distribution of age of non-customers')
```

Age of 28 is quite particular, but we cannot say anything about the difference.

2. Employment type

```python
sorted_counts = df_customers['Employment Type'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, autopct='%1.1f%%')
plt.title('Distribution of employment type of customers')
```

```python
sorted_counts = df_non_customers['Employment Type'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, autopct='%1.1f%%')
plt.title('Distribution of employment type of non-customers')
```

Self-employed people tend to buy insurence

3. Graduation

```python
sorted_counts = df_customers['GraduateOrNot'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, autopct='%1.1f%%')
plt.title('Have customers graduated?')
```

```python
sorted_counts = df_non_customers['GraduateOrNot'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90, counterclock = False, autopct='%1.1f%%')
plt.title('Have non-customers graduated?')
```

Nearly no difference

4. Annual income

```python
df_customers['AnnualIncome'].hist()
plt.title('Annual income of customers')
plt.xlabel('Annual income ($)')
plt.ylabel('count')
print('Average:', df_customers['AnnualIncome'].mean())
print('Median:', df_customers['AnnualIncome'].median())
```

```python
df_non_customers['AnnualIncome'].hist()
plt.xlabel('Annual income ($)')
plt.ylabel('count')
plt.title('Annual income of non-customers')
print('Average:', df_non_customers['AnnualIncome'].mean())
print('Median:', df_non_customers['AnnualIncome'].median())
```

There is an obvious gap between customers and non-customers in annual income. (Use this!)

```python
df_customers['AnnualIncome'].hist(alpha = 0.7, figsize=(10, 6))
df_non_customers['AnnualIncome'].hist(alpha = 0.7)
plt.legend(['Customers','Non-customers'], loc = 2)
plt.xlabel('Annual income ($)')
plt.ylabel('count')
plt.title('Annual income of customers and non-customers')
#plt.figure(figsize=(120, 60))
plt.show()
```
